```python
#Importamos la libreria pandas
import pandas as pd

#utilizando la librería pandas, hacemos la lectura de los datos.'autos.csv'
data_autos=pd.read_csv('autos.csv', sep=(';'))
df=pd.DataFrame(data_autos)
df
```

```python
#Creamos una función que realiza el cambio de un string a un valor entero.
#si el pais es Japan - resultará 1
#si el pais es Europe - resultará 2
#si el pais es USA - resultará 3
def cambio_pais(n):
    if n=='Japan':
        n=1
    elif n=='Europe':
        n=2
    elif n=='USA':
        n=3
    else:
        print("No existe")
#con la palabra return hacemos que la funcion retorne n, eso quiere decir retorna los valores de 1,2 y 3 segun sea el caso.
    return n

#Creamos una nueva columna 'Numero_pais' en el dataframe, y aplicamos la funcion 'cambio_pais' a la columna 'Pais_Origen'.
df['Numero_Pais']=df['Pais_Origen'].apply(cambio_pais)
df

```
