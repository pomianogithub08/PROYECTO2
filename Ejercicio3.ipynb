```python
#Importamos las librería pandas y numpy
import pandas as pd
import numpy as np
#utilizando la librería pandas, hacemos la lectura de los datos.'autos.csv'
data_autos=pd.read_csv('autos.csv', sep=(';'))
df=pd.DataFrame(data_autos)
#Creamos un nuevo dataframe con los datos de Aaeleracion
df_1=pd.DataFrame(df['Aceleracion'])
df_1
```

```python
#Calculamos la media de los datos de la columna 'Aceleracion' y lo almacenamos
#en la variable MEDIA de manera sencilla con la libreria pandas.
MEDIA=df['Aceleracion'].mean()

#Calculamos la desviacion estandar de los datos de la columna 'Aceleracion' y lo almacenamos
#en la variable DESVIACION de manera sencilla con la libreria pandas.
DESVIACION=df['Aceleracion'].std()

#Creamos dos variables que van hacer los parámetros límites.L1 Y L2
L1=MEDIA + DESVIACION
L2=MEDIA - DESVIACION

#Definimos la función que toma el valor de los indices. se compara con los
#limites mediabte condicionales y se clasifica si es alta, media o baja
#la aceleracion.
def cualitativa(a):
    if df_1['Aceleracion'][a] > L1:
        c='Alta'
    elif L2 < df_1['Aceleracion'][a] < L1:
        c='Media'
    elif df_1['Aceleracion'][a] < L2:
        c='Baja'
    else:
        print('Valor no válido')
    
    return c

#si lo desarrollamos con la funcion apply saldría directo.

#def cualitativa(a):
#    if a > L1:
#        c='Alta'
#    elif L2 < a < L1:
#        c='Media'
#    elif a < L2:
#        c='Baja'
#    else:
#        print('Valor no válido')
#    return c
    
#df_1['A']=df_1['Aceleracion'].apply(cualitativa)
```

```python
#Usaremos otra forma.
#Creamos una lista vacia. 'variables_cualitativas'
variables_cualitativas=[]
#Rango de los indices(valores)
valores=range(0,len(df_1['Aceleracion']))
#Recorremos todos los valores con el ciclo for, por cada indice, llamamos a la
#función cualitativa , y el resultado se va almacenando en la lista creada.
#almacena en la lista creada.
for n in valores:
    variable=cualitativa(n)
    variables_cualitativas.append(variable)    
```

```python
#Añadimos una columna nueva llamada 'Variables_Cualitativas', donde estaran
#los datos de la lista 'variables_cualitativas'
df_1['Variables_Cualitativas']=variables_cualitativas
df_1
```
