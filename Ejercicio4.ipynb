```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data_autos=pd.read_csv('autos.csv', sep=(';'))
df=pd.DataFrame(data_autos)
df_1=pd.DataFrame(df['Aceleracion'])
df_1

MEDIA=df['Aceleracion'].mean()
DESVIACION=df['Aceleracion'].std()
L1=MEDIA + DESVIACION
L2=MEDIA - DESVIACION

def cualitativa(a):
    if a > L1:
        c='Alta'
    elif L2 < a < L1:
        c='Media'
    elif a < L2:
        c='Baja'
    else:
        print('Valor no válido')
    return c
    
df_1['Tipo']=df_1['Aceleracion'].apply(cualitativa)
df_1
```
