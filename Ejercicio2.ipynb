```python
#Importamos las librería pandas y numpy
import pandas as pd
import numpy as np
#utilizando la librería pandas, hacemos la lectura de los datos.'autos.csv'
data_autos=pd.read_csv('autos.csv', sep=(';'))
df=pd.DataFrame(data_autos)
df
```

```python
#Creamos una lista llamada 'dis' que se alamacenará todos los datos de
#'Desplazamiento' del dataframe.
#lista vacia
dis=[]
#Definimos una función que recorra todo la columna de la columna de
#'Desplazamiento' del dataframe y almacenamos dato por dato en la lista creada.
def desplazamiento(d):
    for i in df['Desplazamiento']:
        dis.append(i)
#retorna la lista ya almacenada con todos los datos        
    return dis
#Aplicamos la funcion creada a la columna 'Desplazamiento' del dataframe.
desplazamiento(df['Desplazamiento'])


#Creamos una lista llamada 'cab' que se alamacenará todos los datos de
#'Caballo_De_Fuerza' del dataframe.
#lista vacia
cab=[]
#Definimos una función que recorra todo la columna de la columna de
#'Caballo_De_Fuerza' del dataframe y almacenamos dato por dato en la lista creada.
def caballo(c):
    for i in df['Caballo_De_Fuerza']:
        cab.append(i)
#retorna la lista ya almacenada con todos los datos. 
    return cab

#Aplicamos la funcion creada a la columna 'Caballo_De_Fuerza' del dataframe.
caballo(df['Caballo_De_Fuerza'])


#Creamos un diccionario con las listas creadas, y los datos serán nuestras
#dos variables 'Var1'=datos de 'Desplazamiento' y 'Var2'='Caballo_De_Fuerza'
dic={'Var1':dis,
    'Var2':cab}

#creamos un dataframe con el diccionario
df_1=pd.DataFrame(dic)
df_1
```

```python
#Generamos dos columnas llamadas 'Var_A' y 'Var_B', dentro de esas columnas
#estarán los resultados solicitados.
df_1['Var_A']=np.log10(df_1['Var1']/df_1['Var2'])
df_1['Var_B']=np.sqrt(df_1['Var1'])*np.exp(df_1['Var2'])/200
df_1
```

```python
#Generamos una columna nueva llamada 'Var_C' que estarán almacenados
#por valores de acuerdo a una cierta condición si los valores de Var1 es mayor
#que Var2 el valor es 5, en caso contrario es 3.
df_1['Var_C']=''
df_1.loc[df_1['Var1']>df_1['Var2'],'Var_C']=5
df_1.loc[df_1['Var1']<=df_1['Var2'],'Var_C']=3
df_1
```

```python
#De otra manera podemos crear una función, en la cual este definida las
#condicionales para el dataframe creado, la condicion es que si Var1 es mayor
#que Var2 el valor es 5 en caso contrario es 3.
def cualitativas(m):
    #Condición
    if df_1['Var1'][m] > df_1['Var2'][m]:
#Si la condicion se da el valor es 5.
        a=5
#Si no se cumple la condicion    
    else:
#el valor es 3
        a=3
    return a

#creamos el rango de los indices
indices=range(0,len(df_1['Var1']))
#Inicializamos con una lista vacia VarC1
VarC1=[]
#Recorremos todos los valores de los indices
for j in indices:
#Por cada indice , llamamos a la función.    
    p=cualitativas(j)
#Por cada indice los valores se van almacenando en la lista vacia.
    VarC1.append(p)
#Creamos una nueva columna Var_C1, donde estaran los datos de la lista creada.
df_1['Var_C1']=VarC1
df_1
```

```python
#Generamos dos columnas llamadas 'Var_D' y 'Var_E', dentro de esas columnas
#estarán los resultados solicitados.

df_1['Var_D']=1/np.log(df_1['Var1']/df_1['Var2'])*100
df_1['Var_E']=df_1['Var2']**2/df_1['Var1']
df_1
```
